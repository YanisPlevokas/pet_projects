```python
df = pd.get_dummies(df_new, prefix=['exp', 'schedule_new'])
```

```python
df.drop(['salary_from', 'salary_to'], axis=1, inplace=True)
```

```python
regressor = DecisionTreeRegressor(random_state=0)
```

```python
cross_val_score(regressor, X, y, cv=10)
```

```python
train_size = int(0.7 * df.shape[0])
```

```python
X_train, X_valid = X.iloc[:train_size, :], X.iloc[train_size:,:]

y_train, y_valid = y.iloc[:train_size], y.iloc[train_size:]
```

```python
regressor.fit(scipy.sparse.csr_matrix(X_train), y_train)
```

```python
train_preds = regressor.predict(scipy.sparse.csr_matrix(X_train))
valid_preds = regressor.predict(scipy.sparse.csr_matrix(X_valid))

print('Ошибка на трейне',mean_absolute_error(y_train, train_preds))
print('Ошибка на тесте',mean_absolute_error(y_valid, valid_preds))
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42, verbose=1)
```

```python
parameters = {
    "max_depth": np.arange(1, 10, 1),
}
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=5)
```

```python
rand_cv = GridSearchCV(estimator=rf, param_grid=parameters, n_jobs=-1)
```

```python
%%time
rand_cv.fit(X_train,y_train)
```
